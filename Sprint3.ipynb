```python
import pandas as pd
import plotly.express as px

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡πà‡∏ß‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤
price_ranges = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-10000', '10000+']

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ä‡πà‡∏ß‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤
for price_range in price_ranges:
    # ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡∏ä‡πà‡∏ß‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤
    if price_range == '10000+':
        df_price_range = df_pay_true[df_pay_true['price'] >= 10000]
    else:
        price_range_min, price_range_max = map(int, price_range.split('-'))
        df_price_range = df_pay_true[(df_pay_true['price'] >= price_range_min) & (df_pay_true['price'] <= price_range_max)]

    # ‡∏£‡∏ß‡∏°‡∏¢‡∏≠‡∏î‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤‡∏ï‡∏≤‡∏°‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤
    df_grouped = df_price_range.groupby('time_range')['price'].sum().reset_index()

    # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏£‡∏≤‡∏Ñ‡∏≤‡∏£‡∏ß‡∏°‡∏à‡∏≤‡∏Å‡∏°‡∏≤‡∏Å‡πÑ‡∏õ‡∏ô‡πâ‡∏≠‡∏¢
    df_grouped_sorted = df_grouped.sort_values(by='price', ascending=False)

    # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡πÅ‡∏ó‡πà‡∏á‡∏î‡πâ‡∏ß‡∏¢ Plotly Express
    fig = px.bar(df_grouped_sorted,
                 x='time_range',  # ‡πÉ‡∏´‡πâ‡πÅ‡∏Å‡∏ô x ‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤
                 y='price',  # ‡πÉ‡∏´‡πâ‡πÅ‡∏Å‡∏ô y ‡πÄ‡∏õ‡πá‡∏ô‡∏¢‡∏≠‡∏î‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏Ñ‡∏≤
                 labels={'time_range': 'Time Range', 'price': 'Total Price'},  # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î label ‡πÉ‡∏´‡πâ‡πÅ‡∏Å‡∏ô x ‡πÅ‡∏•‡∏∞ y
                 title=f'Total Price by time_range (Price Range {price_range})',
                 color='time_range',  # ‡πÉ‡∏ä‡πâ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏™‡∏µ
                 color_discrete_map=color_time  # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏™‡∏µ‡∏ï‡∏≤‡∏°‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÑ‡∏ß‡πâ‡πÉ‡∏ô color_map
    )

    # ‡πÄ‡∏û‡∏¥‡πà‡∏° label ‡πÉ‡∏´‡πâ‡∏Å‡∏±‡∏ö‡πÅ‡∏Å‡∏ô x ‡πÅ‡∏•‡∏∞ y
    fig.update_xaxes(title_text='Time Range')
    fig.update_yaxes(title_text='Total Price')

    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡∏≠‡∏á‡∏Å‡∏£‡∏≤‡∏ü
    fig.update_layout(width=800, height=600)

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏Å‡∏≥‡∏Å‡∏±‡∏ö‡πÉ‡∏ô‡∏Å‡∏£‡∏≤‡∏ü‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÅ‡∏ó‡πà‡∏á
    for bar in fig.data:
        for i, j in enumerate(bar.y):
            text = f'{j:.0f}'  # ‡πÅ‡∏™‡∏î‡∏á‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° 0 ‡∏´‡∏•‡∏±‡∏Å
            fig.add_annotation(
                x=bar.x[i],
                y=j,
                text=text,
                textangle=0,
                showarrow=False,
                font_size=12,
            )

    # ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡∏£‡∏≤‡∏ü
    fig.show()
```
